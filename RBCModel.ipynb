```python
import numpy as np
from numpy.random import normal
from numpy import log, linspace, zeros, interp, sqrt, array, exp, mean, max
```

```python
# Params and Stucture
k0 = 36
z0 = 0
δ = 0.00788
α = 0.36
β = (1/1.04)**(1/4)
υ = 0.25
Φ = 0.4515
u = lambda c, l: log(c) + Φ*(1-l)**(1-1/υ)/(1-1/υ)
F = lambda k,l: k**α*l**(1-α)
f = lambda k,l: F(k, l) - k*(1-δ)
```

```python
# Arrays
ξ = 1 # For oversampling smaller values of state
kGrid = 2
zGrid = 2
lGrid = 2
kGrid = linspace(33**(1/ξ),38**(1/ξ),kGrid)**ξ;
lGrid = linspace(0.3**(1/ξ),0.35**(1/ξ),lGrid)**ξ;
zGrid = np.array([-.007, 0, 0.007])
nK = len(kGrid)
nZ = len(zGrid)
nL = len(lGrid)
```

```python
U = zeros((nK, nZ, nK, nL));
valueRHS = zeros((nK, nZ));
valueLHS = zeros((nK, nZ));
kpPolicy = zeros((nK, nZ));
lPolicy = zeros((nK, nZ));
cPolicy = zeros((nK, nZ));
iter = 1;
```

```python
R = zeros((nK,nZ,nK,nL))
for ik, k in enumerate(kGrid):
  for iz, z in enumerate(zGrid):
    for il, l in enumerate(lGrid):
      c = np.maximum(exp(z)*F(k, l)+(1-δ)*k-kGrid, 1e-6)
      R[ik, iz, :, il] = u(c, l)

np.shape(R)
```
