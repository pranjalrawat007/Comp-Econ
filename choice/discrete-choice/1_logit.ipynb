```python
import numpy as np
X = np.array(range(100))
D = np.array(range(100))

def u(x,d):
    return (x+d)-0.01*(x+d)**2
```

```python
u(X[1],D)
```

```python
u(X[2],D)
```

```python
def e(D):
    return np.array([np.random.gumbel(loc=d-0.01*d**2, scale=1.0, size=1)[0] for d in D])

e(D)
```

```python
u(X[1],D)+e(D)
```

```python
def d(x,e):
    true_utils = u(x,D)+e(D)
    return np.argmax(true_utils), np.max(true_utils)

d(X[1],e)
```

```python
N = 1000
import seaborn as sns
sns.kdeplot([d(X[99],e)[0] for i in range(N)])
sns.kdeplot([d(X[1],e)[0] for i in range(N)])
```
