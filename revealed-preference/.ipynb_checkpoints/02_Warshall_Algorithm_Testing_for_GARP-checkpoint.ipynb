```python
# N obs on N good bundle
import numpy as np
p = np.array([[1,2,3,4],[2,3,4,5], [2,3,4,5], [2,3,4,5], [1,1,2,3]])
q = np.array([[3,4,5,6], [2,3,7,8], [14,4,6,2], [2,4,2,2], [9,8,7,6]])
N = q.shape[0]
M = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        if np.dot(p[i],q[i]) >= np.dot(p[i],q[j]):
            M[i,j] = 1
M
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def show_graph_with_labels(adjacency_matrix, mylabels):
    rows, cols = np.where(adjacency_matrix == 1)
    edges = zip(rows.tolist(), cols.tolist())
    gr = nx.DiGraph(directed=True)
    gr.add_edges_from(edges)
    nx.draw(gr, node_size=500, labels=mylabels, with_labels=True)
    plt.show()

labels = {0:'A', 1:'B', 2:'C'}
show_graph_with_labels(M, labels)
```

```python
# Obtaining the transitive closure

def warshall(M):
    C = M.copy()
    for i in range(N):
        for j in range(N):
            if M[i,j] == 1:
                C[i,j] = 1
            else:
                C[i,j] = np.inf
                
    for k in range(N):
        for i in range(N):
            for j in range(N):
                if C[i,j] >= C[i,k] + C[k,j]:
                    C[i,j] = C[i,k] + C[k,j]
                    
        if k<(N-1):
            k += 1
        else:
            pass

    for i in range(N):
        for j in range(N):
            if C[i,j] < np.inf:
                C[i,j] = 1
            else:
                C[i,j] = 0
    print(C==M)
    
    return C
```

```python
print(M)
warshall(M)
```
