# Arrays and Parameters

```julia
using Optim, Calculus, LinearAlgebra, NLsolve, Plots, StatsBase,  Roots

# Parameters
β = 0.99
δ = 0.0233
α = 1/3
ϕ = 1.74

# Steady state
k_l = ((1/α)*((1/β)+δ-1))^(1/(α-1)) # from Euler on c
r_ss = α*(k_l)^(α-1)+(1-δ)
y_l = k_l^α # Prodn Fn
c_l = y_l - δ*k_l # from Budget
l_ss = (1+ϕ*c_l/((1-α)*(k_l)^α))^-1 # from Euler on l
k_ss = l_ss * k_l
y_ss = k_ss^α * l_ss^(1-α)
print(l_ss, k_ss)

# Arrays
n_ITER = 200
n_K = 50
n_L = 50
n_Z = 2
grid_K = Array(range(9,12,n_K))
grid_Z = Array([0.95, 1.05])
grid_L = Array(range(0.2,0.6,n_L))
P = Array([0.95 0.05; 0.05 0.95])
grid_V = zeros(n_K,n_Z)
grid_V[:, 1]= -range(90,93,n_K)
grid_V[:, 2]= -range(90,93,n_K)
grid_kp = zeros(n_K,n_Z)
grid_l = zeros(n_K,n_Z)
grid_c = zeros(n_K,n_Z); 
```

# Value Function Iteration

```julia
for iter in range(1,n_ITER,n_ITER)
    print(iter)
    for (k_i,k) in enumerate(grid_K)
       for (z_i,z) in enumerate(grid_Z)
            Vmax = -100
            V_best = -100
            kp_best = 9
            l_best = 9
            for (kp_i,kp) in enumerate(grid_K)
                for (l_i, l) in enumerate(grid_L)
                    c = z*k^α*l^(1-α)+(1-δ)*k-kp
                    ymax = z*k^α*l^(1-α)+(1-δ)*k
                    if (c > 0) & (c<=ymax) & (l<1) & (kp<ymax)
                        V = log(c)+ϕ*log(1-l)+β*dot(P[z_i, :],grid_V[kp_i,:])
                        if V >= Vmax
                            Vmax=V
                            V_best=V
                            kp_best=kp
                            l_best=l
                        end
                    end
                end
            end
            grid_V[k_i,z_i]=V_best
            grid_kp[k_i,z_i]=kp_best
            grid_l[k_i,z_i]=l_best
            grid_c[k_i,z_i]=z*k^α*l_best^(1-α)+(1-δ)*k-kp_best
        end
    end
end
```

# Simulation


```julia
function simulateAR1(N)
    path_z = zeros(N)
    path_z[1] = grid_Z[1]
    FN(a, b) = argmin(abs.(a .- b))
    for t in range(1,N-1,N-1)
        i = FN(grid_Z, path_z[Int(t)])
        path_z[Int(t)+1] = sample(grid_Z, ProbabilityWeights(P[Int(i), :]))
    end
    return path_z
end

path_z = simulateAR1(5000)
plot(path_z)
```

```julia
function simulate(path_z, N=5000)
    k0 = k_ss+1
    path_k = zeros(N+1)
    path_k[1] = k0
    path_l = zeros(N)
    path_c = zeros(N)
    path_y = zeros(N)
    path_i = zeros(N)
    FN(a, b) = argmin(abs.(a .- b))
    for t in range(1,N,N)
        k_i = FN(grid_K, path_k[Int(t)]) 
        z_i = FN(grid_Z, path_z[Int(t)])
        path_l[Int(t)] = grid_l[k_i, z_i]
        path_c[Int(t)] = grid_c[k_i, z_i]
        path_k[Int(t)+1] = grid_kp[k_i, z_i]
        path_y[Int(t)] = path_z[Int(t)]*path_k[Int(t)]^α*grid_l[k_i, z_i]^(1-α)
        path_i[Int(t)] = grid_kp[k_i, z_i] - (1-δ)*path_k[Int(t)]
    end
    return path_z, path_l, path_c, path_y, path_k, path_i
end

path_z, path_l, path_c, path_y, path_k, path_i = simulate(path_z);
```

```julia
c = Array(path_c)[1000:4999]
l = Array(path_l)[1000:4999]
k = Array(path_k)[1000:4999]
y = Array(path_y)[1000:4999];
i = Array(path_i)[1000:4999];

using HPFilter
c_cycle = c - HP(c,1600)
k_cycle = k - HP(k,1600)
y_cycle = y - HP(y,1600)
l_cycle = l - HP(l,1600);
i_cycle = i - HP(i,1600);
```

```julia
using Printf
Base.show(io::IO, f::Float64) = @printf(io, "%1.3f ", f)
println("Std Consumption Cycle:" , 100*std(c_cycle)) 
println("Std Investment Cycle:" , 100*std(i_cycle)) 
println("Std Labor Hours Cycle:" , 100*std(l_cycle)) 
println("Std Output Cycle:" , 100*std(y_cycle)) 
println("Std Capital Cycle:" , 100*std(k_cycle)) 
```

```julia
println("Std Consumption-Output Correlation:" , 100*cor(c_cycle, y_cycle)) 
println("Std Investment-Output Cycle:" , 100*cor(i_cycle, y_cycle)) 
println("Std Labor Hours-Output Cycle:" , 100*cor(l_cycle, y_cycle)) 
println("Std Capital-Output Cycle:" ,100*cor(k_cycle, y_cycle)) 
```
