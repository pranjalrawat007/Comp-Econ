### 1-Variable System:
***
$$
y′(t) = b−a y(t)\\
\text{$y(0)$ given}\\
t ∈ (0,T]\\
a,b > 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters
y0 = 10
T_max = 10
T_min = 0.01
a = 0.7
b = -5

# Algorithm Parameters
N = 99

def grids(T_min, T_max, N):
    y_grid = np.linspace(T_min, T_max, N+1)
    t_grid = np.linspace(T_min, T_max, N+1)
    return t_grid, y_grid

def interpolate(t_grid, y_grid, y0, a, N):
    y_grid[0] = y0
    for idx in range(N):
        y_grid[idx+1] = y_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-a * y_grid[idx]+b)
    return y_grid

# Generate Grids
t_grid, y_grid = grids(T_min, T_max, N)
print(t_grid[0], t_grid[-1], len(t_grid))

# Use Finite Differences 
y_grid = interpolate(t_grid, y_grid, y0, a, N)

plt.plot(t_grid, interpolate(t_grid, y_grid, y0, a, N))
plt.title('y(t) vs t')
plt.show()
```

```python
# Phase Diagram
y_grid = np.linspace(-10,10,100)
yp_grid = -a*y_grid+b
plt.plot(yp_grid, y_grid)
plt.axhline(y=0, color='r', linestyle='-')
plt.axvline(x=0, color='r', linestyle='-')
plt.title("Phase Diagram: y'(t) vs y(t)")
plt.show()
```

### 2-Variable System:
***
$$
y′(t) = - a y(t) - b x(t) \\
x′(t) = - b y(t) - a x(t) \\
\text{$y(0), x(0)$ given}\\
t ∈ (0,T]\\
a,b > 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters
y0 = 10
x0 = 15
T_max = 10
T_min = 0.01
a = 0.7
b = 0.2

# Algorithm Parameters
N = 999

def grids(T_min, T_max, N):
    x_grid = np.linspace(T_min, T_max, N+1)
    y_grid = np.linspace(T_min, T_max, N+1)
    t_grid = np.linspace(T_min, T_max, N+1)
    return t_grid, y_grid, x_grid

def interpolate(t_grid, y_grid, x_grid, y0, x0, a, b, N):
    y_grid[0] = y0
    x_grid[0] = x0
    for idx in range(N):
            y_grid[idx+1] = y_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-a * y_grid[idx] - b * x_grid[idx])
            x_grid[idx+1] = x_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-b * y_grid[idx] - a * x_grid[idx])
    return y_grid, x_grid

# Generate Grids
t_grid, y_grid, x_grid = grids(T_min, T_max, N)
print(t_grid[0], t_grid[-1], len(t_grid))

# Use Finite Differences 
y_grid, x_grid = interpolate(t_grid, y_grid, x_grid, y0, x0, a, b, N)
plt.plot(t_grid, x_grid, label='x(t)')
plt.plot(t_grid, y_grid, label='y(t)')
plt.legend()
plt.show()
```
