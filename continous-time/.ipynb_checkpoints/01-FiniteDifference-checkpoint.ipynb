Exponential Decay is a First Order ODE:\
$$
y′(t) = −a y(t)\\
\text{$y(0)$ given}\\
t ∈ (0,T]\\
a > 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters
y0 = 10
T_max = 10
T_min = 0.01
a = 0.7

# Algorithm Parameters
N = 99

def grids(T_min, T_max, N):
    y_grid = np.linspace(T_min, T_max, N+1)
    t_grid = np.linspace(T_min, T_max, N+1)
    return t_grid, y_grid

def interpolate(t_grid, y_grid, y0, a, N, method='forward'):
    y_grid[0] = y0
    for idx in range(N):
        if method == 'forward':
            y_grid[idx+1] = y_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-a * y_grid[idx])
        elif method == 'backward':
            y_grid[idx+1] = y_grid[idx] * (1/(1+a*(t_grid[idx+1]-t_grid[idx])))
        elif method == 'crank-nikolson':
            y_grid[idx+1] = y_grid[idx] * (1 - 0.5*a*(t_grid[idx+1]-t_grid[idx]))/(1+0.5*a*(t_grid[idx+1]-t_grid[idx]))
    return y_grid

# Generate Grids
t_grid, y_grid = grids(T_min, T_max, N)
print(t_grid[0], t_grid[-1], len(t_grid))

# Use Finite Differences 
y_grid = interpolate(t_grid, y_grid, y0, a, N, method = 'crank-nikolson')

for i in ['forward', 'backward', 'crank-nikolson']:
    plt.plot(t_grid, interpolate(t_grid, y_grid, y0, a, N, method = i), label=i)
plt.legend()
plt.title('y(t) vs t')
plt.show()
```

2-Variable Exponential Decay is a First Order ODE System:\
$$
y′(t) = - a y(t) - b x(t) \\
x′(t) = - b y(t) - a x(t) \\
\text{$y(0), x(0)$ given}\\
t ∈ (0,T]\\
a,b > 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters
y0 = 10
x0 = 15
T_max = 10
T_min = 0.01
a = 0.7
b = 0.2

# Algorithm Parameters
N = 999

def grids(T_min, T_max, N):
    x_grid = np.linspace(T_min, T_max, N+1)
    y_grid = np.linspace(T_min, T_max, N+1)
    t_grid = np.linspace(T_min, T_max, N+1)
    return t_grid, y_grid, x_grid

def interpolate(t_grid, y_grid, x_grid, y0, x0, a, b, N):
    y_grid[0] = y0
    x_grid[0] = x0
    for idx in range(N):
            y_grid[idx+1] = y_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-a * y_grid[idx] - b * x_grid[idx])
            x_grid[idx+1] = x_grid[idx] + (t_grid[idx+1]-t_grid[idx]) * (-b * y_grid[idx] - a * x_grid[idx])
    return y_grid, x_grid

# Generate Grids
t_grid, y_grid, x_grid = grids(T_min, T_max, N)
print(t_grid[0], t_grid[-1], len(t_grid))

# Use Finite Differences 
y_grid, x_grid = interpolate(t_grid, y_grid, x_grid, y0, x0, a, b, N)
plt.plot(t_grid, x_grid, label='x(t)')
plt.plot(t_grid, y_grid, label='y(t)')
plt.legend()
plt.show()
```

K-Variable First Order ODE System:\
$$
X′(t) = A X(t)\\
\text{$X(0)$ given}\\
t ∈ (0,T]\\
\text{$A<0$ and has eigenvalues in unit circle.}\\
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters
T_max = 10
T_min = 0.01
K = 7 # number of variables

# Generate Initial States (randomly)
X0 = np.random.binomial(20, 0.5, K)

# Generate A-Matrix (randomly)
A = np.identity(K)
A = A + np.random.uniform(low=0., high=.25, size=(K, K))
A = - A / A.sum(axis=1, keepdims=1)
print(A)

# Algorithm Parameters
N = 999

def grids(T_min, T_max, N, X0):
    X_grid = np.zeros((N+1, X0.shape[0]))
    t_grid = np.linspace(T_min, T_max, N+1)
    return t_grid, X_grid

def interpolate(t_grid, X_grid, X0, A, N):
    X_grid[0, :] = X0
    for i in range(N):
        for j in range(X0.shape[0]):
            #print(i,j)
            X_grid[i+1, j] = X_grid[i,j] + (t_grid[i+1]-t_grid[i]) * np.dot(A[j], X_grid[i,:])
    return X_grid

# Generate Grids
t_grid, X_grid = grids(T_min, T_max, N, X0)
print(t_grid[0], t_grid[-1], len(t_grid))

# Use Finite Differences 
X_grid = interpolate(t_grid, X_grid, X0, A, N)
for i in range(X0.shape[0]):
    plt.plot(t_grid, X_grid[:,i], label=f'x{i}(t)')
plt.legend()
plt.show()
```
